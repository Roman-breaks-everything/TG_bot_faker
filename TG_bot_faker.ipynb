```python
# подключение библиотек
import json
from secrets import token_urlsafe
from faker import Faker
from telebot import TeleBot, types

# TODO: вставить свой токен
TOKEN = 'INSERT_YOUR_TOKEN_HERE'
bot = TeleBot(TOKEN, parse_mode='html')
# библиотека для генерации тестовых ФИО
# указываем язык - русский
faker = Faker('ru_RU')

# обработчик команды '/start'
@bot.message_handler(commands=['start'])
def start_message_handler(message: types.Message):
    # отправляем ответ на команду '/start'
    bot.send_message(
        chat_id=message.chat.id,
        text="Введите нужное количество пользователей, максимум - 10."
    )

# обработчик всех остальных сообщений
@bot.message_handler()
def message_handler(message: types.Message):
    # определяем количество тестовых пользователей
    # или отправляем ошибку
    payload_len = 0
    try:
        payload_len = int(message.text)
        if payload_len <= 0:
            raise ValueError
    except ValueError:
        bot.send_message(chat_id=message.chat.id, text="Хорошая попытка! Введите число больше 0.")
        return
        bot.send_message(
        chat_id=message.chat.id,
        text="Если нужны еще данные, введите нужное количество пользователей, максимум - 10",
        reply_markup=main_menu_reply_markup
    )

    # генерируем тестовые данные для выбранного количества пользователей
    # при помощи метода simple_profile
    total_payload = []
    for _ in range(payload_len):
        user_info = faker.simple_profile()
        user_info['ФИО'] = user_info.pop('name')
        user_info['Дата рождения'] = user_info.pop('birthdate')
        user_info['Адрес'] = user_info.pop('address')
        user_info['Телефон'] = f'+7{faker.msisdn()[3:]}'
        user_info['Почта'] = user_info.pop('mail')
        user_info['Пароль'] = token_urlsafe(10)
        user_info['Пол'] = user_info.pop('sex')
        user_info['Имя аккаунта'] = user_info.pop('username')
        total_payload.append(user_info)

    # сериализуем данные в строку
    payload_str = json.dumps(
        obj=total_payload,
        indent=2,
        sort_keys=True,
        ensure_ascii=False,
        default=str
    )

    # отправляем результат
    bot.send_message(
        chat_id=message.chat.id,
        text=f"Данные для {payload_len} пользователей:\n<code>"
            f"{payload_str}</code>"
    )

# главная функция программы
def main():
    # запускаем нашего бота
    bot.infinity_polling()

if __name__ == '__main__':
    main()
```
